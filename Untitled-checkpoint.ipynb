```python
import tensorflow as tf
from tensorflow.keras.layers import DepthwiseConv2D as KerasDepthwiseConv2D
from tensorflow.keras import layers
from tensorflow.keras.models import model_from_json
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Decorate the custom class to make it serializable
@tf.keras.utils.register_keras_serializable(package='Custom')
class CustomDepthwiseConv2D(KerasDepthwiseConv2D):
    def __init__(self, groups=1, **kwargs):
        kwargs.pop('groups', None)  # Remove 'groups' from kwargs
        super().__init__(**kwargs)
        self.groups = groups  # Save 'groups' for custom behavior

    def build(self, input_shape):
        if self.groups != 1:
            raise ValueError(f"groups = {self.groups} not supported by CustomDepthwiseConv2D")
        super().build(input_shape)

# Custom Model Class if you're using a custom model (Functional API model)
@tf.keras.utils.register_keras_serializable(package='Custom')
class CustomModel(tf.keras.Model):
    def __init__(self, layers, **kwargs):
        super().__init__(**kwargs)
        self.layers = layers

    def call(self, inputs):
        x = inputs
        for layer in self.layers:
            x = layer(x)
        return x

# Load Haar Cascade for Face Detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Load Anti-Spoofing Model JSON and Weights
json_path = "mobilenet_face_antispoofing_model.json"
weights_path = "finalyearproject_antispoofing_model_98-0.942647.h5"

# Check if files exist
if not os.path.exists(json_path):
    print(" Model JSON file not found.")
else:
    print(" Model JSON file found.")

if not os.path.exists(weights_path):
    print(" Model weights file not found.")
else:
    print(" Model weights file found.")

# Register custom objects and layers
custom_objects = {
    'DepthwiseConv2D': CustomDepthwiseConv2D,
    'Functional': tf.keras.Model,  # Register Functional model class if used
    'CustomModel': CustomModel  # Register any custom model class if necessary
}

try:
    # Load model architecture from JSON file
    with open(json_path, "r") as json_file:
        model_json = json_file.read()

    # Load model with custom objects
    model = model_from_json(model_json, custom_objects=custom_objects)
    model.load_weights(weights_path)
    print(" Model loaded successfully with custom layers.")

except Exception as e:
    print(f" Error with loading model: {e}")

# Open Webcam
video = cv2.VideoCapture(0)

def preprocess_face(face):
    """ Resize, normalize, and prepare face for model prediction. """
    resized_face = cv2.resize(face, (160, 160))
    resized_face = resized_face.astype("float32") / 255.0
    resized_face = np.expand_dims(resized_face, axis=0)
    return resized_face

# Live Camera Feed in Jupyter Notebook
frame_count = 0  # Counter to refresh display
try:
    while True:
        ret, frame = video.read()
        if not ret:
            print(" Error: Failed to capture frame.")
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)

        for (x, y, w, h) in faces:
            face = frame[y-5:y+h+5, x-5:x+w+5]
            processed_face = preprocess_face(face)
            
            if model:
                preds = model.predict(processed_face)[0]
                label = "Real" if preds < 0.5 else "Digital Image"
                color = (0, 255, 0) if label == "Real" else (0, 0, 255)
                
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        
        # Display in Jupyter Notebook
        if frame_count % 5 == 0:  # Refresh every 5 frames to reduce flickering
            clear_output(wait=True)
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.axis("off")
            display(plt.gcf())
        
        frame_count += 1
        time.sleep(0.1)  # Small delay for smoother display

except KeyboardInterrupt:
    print(" Stopping camera...")

finally:
    video.release()
    print(" Video capture released.")
```
